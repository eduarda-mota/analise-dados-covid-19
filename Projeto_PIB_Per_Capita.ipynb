```python
# Importação das bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Leitura dos dados
dados = pd.read_excel('Dados_Pib.xlsx')
```

```python
dados.head(10)
```

**Evolução da Renda Per Capita por Estado**

```python
# Sistema de Grids

# Cor de fundo
cor_fundo = '#f5f5f5'

# Criação do sistema de Grids
grids = sns.FacetGrid(dados, col='Territorialidades', hue='Territorialidades', col_wrap=4)

# Gráfico de linha
grids = grids.map(plt.plot, 'Ano', 'PIB per capita')

# Sombra + títulos dos gráficos
grids = grids.map(plt.fill_between,  'Ano', 'PIB per capita', alpha=0.2).set_titles('{col_name} Territorialidades')

# Ajuste dos títulos
grids = grids.set_titles('{col_name}')

# Título principal
grids = grids.fig.suptitle('Evolução da Renda Per Capita por Estado', fontsize=18)

# Ajuste do título principal
plt.subplots_adjust(top=0.92);
```
