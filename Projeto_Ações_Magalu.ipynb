```python
# Importação das bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import warnings
warnings.filterwarnings('ignore')
```

```python
# Leitura dos dados
dados = pd.read_excel('Vase_004 - Magalu - Sem Resolução.xlsx')
```

```python
# Dimensão dos dados
dados.shape
```

```python
dados.head()
```

```python
# Informações
dados.info()
```

```python
# Estatística
dados.describe()
```

```python
# Colocando a coluna Data como index
novosdados = dados.set_index('Data')
novosdados.head()
```

**Análise do valor das ações da Magalu (fechamento) no ano de 2021**

```python
# Série Temporal
plt.figure(figsize=(16, 5))
plt.style.use('seaborn-darkgrid')
plt.title('Análise das Ações da Magalu - Fechamento', fontsize=15, loc='left')
plt.plot(novosdados.index, novosdados['Fechamento'])

plt.xlabel('Período da Cotação')
plt.ylabel('Valor da Ação (R$)');
```

```python
# Últimos registros
novosdados.tail()
```

```python
# Adicionando uma média móvel de 5 e de 30 dias à série temporal
media_movel = novosdados['Fechamento'].rolling(5).mean()
media_tendencia = novosdados['Fechamento'].rolling(30).mean()

plt.figure(figsize=(16, 5))
plt.style.use('seaborn-darkgrid')
plt.title('Análise das ações da magalu - Fechamento', fontsize=15, loc='left')

plt.plot(novosdados.index, novosdados['Fechamento'])
plt.plot(novosdados.index, media_movel)
plt.plot(novosdados.index, media_tendencia)

plt.xlabel('Período da Cotação')
plt.ylabel('Valor da Ação (R$)');
```

Em meados de Agosto, houve uma queda significante no valor das ações da Magalu.

**Análise mensal do valor das ações da Magalu (2021)**

```python
# Boxplot Mensal

dados['Mes'] = dados['Data'].dt.month

plt.figure(figsize=(16, 5))
sns.boxplot(data=dados, x='Mes', y='Fechamento', color='purple');
```

Percebe-se que o mês em que houve uma maior oscilação do valor das ações foi o mês de Março, em contrapartida o mês de menor oscilação foi o mês de Junho.

```python
# Análise Estatística
dados.groupby(['Mes']).describe()['Fechamento']
```

**Visão Geral**

```python
Grafico = go.Figure(
    data=[
          go.Candlestick(
              x= novosdados.index,
              open = novosdados['Abertura'],
              high = novosdados['Maior'],
              low = novosdados['Menor'],
              close = novosdados['Fechamento'],
          )
    ]
)

Grafico.update_layout( xaxis_rangeslider_visible=False )
Grafico.show()
```

Com o gráfico acima, pode-se afirmar que houve um maior número de fechamentos em queda do que em alta.
