A análise das ações das empresas de energia (de 2021 a 2022) foi realizada utilizando dados obtidos de uma planilha do Excel.

```python
# Importação das bibliotecas
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Leitura dos dados
dados = pd.read_excel('Dados_Empresas_Energia.xlsx')
```

```python
dados.head()
```

```python
dados.set_index('Data', inplace=True)
dados.head()
```

**Analisando as ações das empresas de energia**

```python
# Gráfico

# Estilo do gráfico
plt.style.use('seaborn-darkgrid')

# Tamanho da figura
plt.figure(figsize=(16, 7))

# Título
plt.title('Análise das Ações das Empresas de Energia', loc='center', fontsize=19, fontweight=0.2)

# Plot da Petrobrás
plt.plot(dados.index, dados['Petrobras'], color='#008c4a', linewidth=3, alpha=0.7)
# Texto da Petrobrás
plt.text(dados.index[-1], dados['Petrobras'].tail(1), 'Petrobrás', color='#008c4a', size='large', horizontalalignment='left')

# Plot das demais empresas
for coluna in dados.columns[1:]:
  plt.plot(dados.index, dados[coluna], color='gray', linewidth=2, alpha=0.7)
  plt.text(dados.index[-1], dados[coluna].tail(1), coluna, color='gray', size='large', horizontalalignment='left')

# Labels
plt.xlabel('Período')
plt.ylabel('Preço de Fechamento (R$)');
```
