A análise da relação entre a renda e o tempo de experiência foi realizada utilizando a base de dados obtida no site Kaggle.

https://www.kaggle.com/datasets/karthickveerakumar/salary-data-simple-linear-regression?datasetId=10624&sortBy=voteCount

```python
# Importação das bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Leitura dos dados
dados = pd.read_csv('Salary_Data.csv')
dados.head()
```

```python
# Renomeação das colunas
dados.rename(columns={
    'YearsExperience' : 'Xp',
    'Salary' : 'Renda'
}, inplace=True)
```

```python
# Dimensão dos dados
dados.shape
```

```python
# Campos nulos
dados.isnull().sum()
```

```python
# Estatística
dados.describe()
```

**Análise da Renda**

```python
plt.figure(figsize=(10, 6))
plt.title('Análise da Renda')
sns.kdeplot(dados['Renda'], color='green');
```

```python
plt.figure(figsize=(4, 4))
sns.boxplot(dados['Renda'], color='purple');
```

**Análise do Tempo de Experiência**

```python
plt.figure(figsize=(10, 6))
plt.title('Análise do Tempo de Experiência')
sns.kdeplot(dados['Xp'], color='blue');
```

```python
plt.figure(figsize=(4, 4))
sns.boxplot(dados['Xp'], color='orange');
```

**Relação entre a renda e o tempo de experiência**

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(data=dados, x='Renda', y='Xp');
```

```python
plt.figure(figsize=(10, 5))
sns.regplot(data=dados, x='Renda', y='Xp');
```

Percebe-se que há uma relação de linearidade entre as duas variáveis. Quanto maior o tempo de experiência, maior é a renda.

**Correlação**

```python
correlacao = dados.corr('pearson')
plt.figure(figsize=(10, 5))
sns.heatmap(correlacao, annot=True, cbar=True);
```

```python
correlacao = dados.corr('spearman')
plt.figure(figsize=(10, 5))
sns.heatmap(correlacao, annot=True, cbar=True);
```

Por fim, observa-se que a correlação entre as duas variáveis é forte e positiva, ou seja, as duas variáveis tendem a aumentar e diminuir juntas, se relacionando de forma linear.
